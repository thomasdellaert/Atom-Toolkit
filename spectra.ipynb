```python
from atomtoolkit import IO
import matplotlib.pyplot as plt
from atomtoolkit import *
import atomtoolkit.render.lineshapes as lineshapes
from atomtoolkit.atom import Atom, Transition
import atomtoolkit.render.spectra as spectra
from pprint import pprint
import numpy as np
```

```python
a = Atom.load('species/Yb_II_173.atom')
a.B = Q_(4.6, 'G')
line369 = a.transitions[('4f14.6s 2S1/2', '4f14.6p 2P*1/2')]
line935 = a.transitions[('4f14.5d 2D3/2', '4f13.(2F*<7/2>).5d.6s.(3D) 3[3/2]*1/2')]
line411 = a.transitions[('4f14.6s 2S1/2', '4f14.5d 2D5/2')]
line760 = a.transitions[('4f13.(2F*).6s2 2F*7/2', '4f13.(2F*<7/2>).5d.6s.(1D) 1[3/2]*3/2')]
```

```python
def add_ticks():
    ax = plt.gca()
    ax.grid(True, 'major', axis='x')
    ax.minorticks_on()
    ax.grid(True, 'minor', axis='x', linestyle=':', linewidth=1)
```

```python
%matplotlib inline
plt.figure(figsize=(20, 9))

lineshape = lineshapes.LorentzianLineShape(gamma=.1)
mod_lineshape = lineshapes.ModLorentzianLineShape(gamma=.1, depth=1, mod_freq_GHz=.048)

spectra.plot_hyperfine_spectrum(line935, lineshape=lineshape, coloring='l', num_points=1000)

add_ticks()
plt.xlim(3.205e5+67, 3.205e5+72.5)

plt.show()
```

```python
%matplotlib inline

plt.figure(figsize=(20, 9))

lineshape = lineshapes.LorentzianLineShape(.0005)
mod_lineshape = lineshapes.ModLorentzianLineShape(gamma=.002, depth=.6, mod_freq_GHz=0.034)

spectra.plot_zeeman_spectrum(line369["F=2", "F=3"],
                       lineshape=lineshape, coloring='l', num_points=1000)
add_ticks()

plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(20, 9))

lineshape = lineshapes.LorentzianLineShape(gamma=.0003)
mod_lineshape = lineshapes.ModLorentzianLineShape(gamma=.0005, depth=.5, mod_freq_GHz=0.0002)

spectra.plot_zeeman_spectrum(line411[("F=3", "F=1")],
                       lineshape=lineshape, coloring='transition_type', num_points=1000)
add_ticks()

plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(20, 10))

lineshape = lineshapes.LorentzianLineShape(gamma=.1)
mod_lineshape = lineshapes.ModLorentzianLineShape(gamma=.01, depth=.2, mod_freq_GHz=.048)

spectra.plot_hyperfine_spectrum(line760,
                       lineshape=lineshape, coloring='u', num_points=1000)
add_ticks()

plt.ylim(0, 1)
plt.xlim(3.944e5+17.5, 3.944e5+32.5)

plt.show()
```
